what I understood:
1. isa:Paragraph:Clause is read such that the last word representing 
the head entity is followed by the first word representing 
edge/relation that connects to the second word which is the tail entity
e.g. "Clause is a paragraph"

2. "isa" is another alias for "subclass of" e.g. "Clause subclass of Paragraph"

3. The fragments title(STRING), last_updated(STRING), first_trial_date(DATE)
succeeding the concepts like Judgement, Gazette etc. are these entities properties 
with their provided data type
